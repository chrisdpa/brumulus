### Describe our system

```python
temp_err = np.arange( -10, 10, 0.1)
temp_change = np.arange( -1, 1, 0.05 )
chill_out = np.arange( 0, 100, 1 )

# How close to the target temperature are we
te_too_cold = fuzz.trapmf(temp_err, [-10, -10,-1.5,-1])
te_cold     = fuzz.trimf(temp_err, [-1.5,-1,-0.5])
te_optimal  = fuzz.trimf(temp_err, [-0.5,0,0.5])
te_hot      = fuzz.trimf(temp_err, [0.5,1,1.5])
te_too_hot  = fuzz.trapmf(temp_err, [1,1.5,10,10])

tc_cooling = fuzz.trimf( temp_change, [-1,-1,-0.25])
tc_steady  = fuzz.trimf( temp_change, [-0.25,0,0.25])
tc_heating = fuzz.trimf( temp_change, [0.25,1,1])


# Output - Chiller
co_off    = fuzz.trimf(chill_out, [0,0,5])
co_low    = fuzz.trimf(chill_out, [0,20,40])
co_medium = fuzz.trimf(chill_out, [20,40,60])
co_high   = fuzz.trapmf(chill_out, [40,60,100,100])


```

```python
def temp_error_category( set_point = 20, current = 20 ):
    te_current = current - set_point
    te_too_cold_cat = fuzz.interp_membership( temp_err, te_too_cold, te_current)
    te_cold_cat     = fuzz.interp_membership( temp_err, te_cold, te_current)
    te_optimal_cat  = fuzz.interp_membership( temp_err, te_optimal, te_current)
    te_hot_cat      = fuzz.interp_membership( temp_err, te_hot, te_current)
    te_too_hot_cat  = fuzz.interp_membership( temp_err, te_too_hot, te_current)
    return dict( too_cold = te_too_cold_cat, 
                 cold = te_cold_cat, 
                 optimal = te_optimal_cat,
                 hot = te_hot_cat,
                 too_hot = te_too_hot_cat)

temp_err_in = temp_error_category( 20, 22 )
print "Temp Error", temp_err_in
```

```python
#If temp is ?
rule_temp_optimal = temp_err_in['optimal']
rule_temp_hot     = temp_err_in['hot']
rule_temp_too_hot = temp_err_in['too_hot']

#Then:
when_optimal = np.fmin( rule_temp_optimal, co_off )
when_hot     = np.fmin( rule_temp_hot, co_low )
when_too_hot = np.fmin( rule_temp_too_hot, co_medium )

aggregate_membership = np.fmax( when_optimal, np.fmax( when_hot, when_too_hot ) )
result = fuzz.defuzz( chill_out, aggregate_membership, 'centroid' )
print result
print result/100*30
```
